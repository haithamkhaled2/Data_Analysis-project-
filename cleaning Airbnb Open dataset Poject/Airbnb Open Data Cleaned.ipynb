```python
import pandas as pd 
df = pd.read_csv(r"C:\Users\haith\Downloads\Airbnb_Open_Data.csv")
df
```

```python
df.columns
```

```python
columns_to_keep = ['id', 'NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',]
```

```python
df = df[columns_to_keep]
df.shape
```

```python
df.drop(columns='id', inplace=True)
df
```

```python
modified_column=[]
for i in df.columns: 
    modified_column.append(i.upper()) 
modified_column
```

```python
df.columns = modified_column
df
```

```python
df.duplicated().sum()
df.duplicated().value_counts()
```

```python
df= df.drop_duplicates()
df.duplicated().value_counts()
```

```python
df.isna().sum()
df.dropna(inplace=True)
df.isna().sum()
```

```python
df['HOST_IDENTITY_VERIFIED']=df['HOST_IDENTITY_VERIFIED'].str.upper()
df
```

```python
df['INSTANT_BOOKABLE']=df['INSTANT_BOOKABLE'].apply(lambda x: 'Available' if x == True else 'Not Available')
df
```

```python
df.reset_index(drop=True , inplace=True)
df
```

```python
df[['PRICE','SERVICE FEE']].replace('\$','', regex=True, inplace=True)
df[['PRICE','SERVICE FEE']].replace(',','', regex=True, inplace=True)
df[['PRICE','SERVICE FEE']] = df[['PRICE','SERVICE FEE']].replace('\$','', regex=True, )
df[['PRICE','SERVICE FEE']] = df[['PRICE','SERVICE FEE']].replace(',','', regex=True, )
df
```

```python
df[['PRICE','SERVICE FEE']]=df[['PRICE','SERVICE FEE']].astype(int)
df[['PRICE','SERVICE FEE']].dtypes
```

```python
df.to_csv(r"C:\Users\haith\PyFolder\Airbnb_Open_Data_Cleaned.csv", index=False)
```
