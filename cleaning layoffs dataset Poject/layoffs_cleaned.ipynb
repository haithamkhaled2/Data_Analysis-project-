```python
import pandas as pd 

df = pd.read_csv(r"C:\Users\haith\Downloads\layoffs.csv")
df.head(10)
```

```python

df.filter(items=['company', 'location']).apply(lambda x: x.str.strip())
```

```python
df.loc[df['industry'].str.startswith('Crypto', na=False), 'industry'] = 'Crypto'
df['industry'].unique()
```

```python
l2 = df[df['industry'].notna()]
l1 = df.copy()
merged = l1.merge(l2[['company','industry']], how='left', on=['company'], suffixes=('', '_l2'))
merged['industry'] = merged['industry'].fillna(merged['industry_l2'])
df['industry'] = merged['industry'] 
```

```python
df.drop(df[df[['total_laid_off', 'percentage_laid_off']].isnull().all(axis=1)].index,inplace=True)
```

```python
df['date'].dtype
```

```python
df['date']=df['date'].astype(str)
df['date'].head(10)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date']= df['date'].dt.strftime('%Y-%m-%d')
df
```

```python
df.reset_index(drop=True, inplace=True)
df.fillna('', inplace=True)
#df.drop_duplicates(subset='date', inplace=True)
df
```

```python
df['funds_raised_millions'] = df['funds_raised_millions'].replace('', 0).astype(float)
df['status'] =df['funds_raised_millions'].apply(lambda x: 'high funds' if x >= 50 else 'low funds')
df
```
